```python
l = [1,2,3,4,5] # squre of each
def square(l) : 
    s =[]    
    for i in l : 
        s.append(i**2)
    return s
square(l)
```

```python
def square_ele(x) : 
    return x**2
```

```python
l = [1,2,3,4,5,6,7,8]
```

```python
# map( func , iterable )
```

```python

list(map(square_ele , l))
```

```python
def add(x) : 
    return x + 10
```

```python
l = [10,20,30,40,50]
```

```python
list(map(add,l))
```

```python
# Using lambda with
list(map(lambda x : x**2, [1,2,3,4,5]))
```

```python
list(map(lambda x : x+10, [1,2,3,4,5]))
```

```python
l = ["-1" , "-2" , "-3" , "-4"]
```

```python
list(map(lambda x : int(x) , l))
```

```python
l1 = [100,200,300,400,500]
l2 = [1,2,3,4,5]
#add both the list
list(map(lambda x,y : x+y , l1,l2))
```

```python
s = "datascience" # convert to upercase
p = list(map(lambda s : s.upper() , s))
"".join(p)
```

```python
words = ["ajay" , "shubham"]
list(map(str.capitalize , words))
```

```python
# Grades to cgpa
grade = ['A' , 'B' , "C" ,'D']
list(map(lambda g : 4 if g=="A" else (3 if g=="B" else 2) , grade))
```
