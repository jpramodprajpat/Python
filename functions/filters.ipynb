```python
# filter function>> it is used to filter a iterable based on some condition
# syntax >> filter(func , iterable)
```

```python
l = [1,2,2,5,6,7,8,10]
```

```python
list(filter(lambda x : x%2== 0 , l))
```

```python
list(filter(lambda x : x%2 != 0 , l))
```

```python
list(filter(lambda x : x>5 ,l))
```

```python
def check_even(n) : 
    if(n%2==0) : 
        return True
    return False
```

```python
num = [1,2,3,4,6,8,12,19]
list(filter(check_even,num))
```
