```python
# take input as list and return all numeric values
data = [1,2,"pramod " , 12.3,12,"koom",10 ,[12,90,11,"mahir",5]]

def numeric(data):
     n = []
     for i in data : 
        if type(i) == int or type(i) == float :
            n.append(i)
        elif type(i) == list : 
            for s in i : 
                if type(s) == int or type(s) == float :
                  n.append(s)
     return n

print(numeric(data))
```

```python
print(numeric(data))
```

```python
# variable length Argument in Functions --> When you dont know the number of Argument we use (*)
```

```python
#let a function takes some numbers and return the sum
def sum1(*args) : 
    s = 0 
    for i in args : s+= i
    return s
```

```python
sum1(1,2,30,4,5)
```

```python
def test(*arg,a) : 
    return arg, a
```

```python
test(1,2,4,5, a="ram") # this is very important for default arguement (a = ?) otherwise it will show errror ,and rest 1,2,4,5 become part of *arg
```

```python
# Varible argument -->> takes any no of arguments
# Positional argumet --->> default arguments
```

```python
# Q) write a function which can take any values and return only list among them in the list
def return_list(*arg) : 
    l = []
    for i in arg : 
        if type(i) == list : l.append(i)
    return l
```

```python
return_list(1,2,4, 8+8j, {1,2,3} , True , [1,2,3] , [1,2,[5,6]])
```

```python
# Varible length argumet and  Keyword Argument Together
def test1(*args1, **args2) : 
    return args1 , args2
```

```python
result = test1(10, 20, 30, name="Pramod", city="Bhavnagar")
print(result)
```
