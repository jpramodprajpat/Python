```python
# Generator Function
```

```python
# Regular Function  >> takes a list and return a squre of each
def squre(n) : return [n*n for n in n]

squre([0,1,2,3,4,5])
```

```python
# Regular function
def square(n) : 
    result = []
    for i in range(n) : result.append(i*i)
    return result
square(10)
```

```python
# regular function everything return in one go >> store all -> memory useage💹 and compute all together -> excetution time 💹
# not good for the tasks for big computation
# can we have something that Genreate result one by one instead of one go  : Yes Generative function
  # Regular function : uses return statement
  # Generative funciton : uses yield 
     # yeild returns >> genreative object which is very similar to iterator object and then we can use next() to access further
```

```python
# fibonacci serise
def gen_fibo(n) : 
    a = 0
    b = 1
    for i in range(n) : # 0,1,2,3,4,---,n-1
        yield a
        a , b = b , a+b
```

```python
fibo = gen_fibo(10000000)
```

```python
next(fibo)
```
