```python
# Reduce >> folding/Reduction
# Syntax >> reduce(func, Iterable)
# Reduce will alwage take two argument function and Iterable should not be empty
from functools import reduce
```

```python
l = [1 , 2, 2, 3, 4, 5,  6]
reduce(lambda x,y : x+y , l)
```

```python
# if you take function with more than two arguments 
l = [1 , 2, 2, 3, 4, 5,  6]
reduce(lambda x,y,z : x+y+z ,l)
```

```python
l = [1 , 2, 2, 3, 4, 5,  6]
reduce(lambda x,y : x*y , l)
```

```python
words = ["I" , "am" , "Pramod" ,"learning" , "Engineer"]
reduce(lambda x,y : x +" " + y ,words)
```

```python
# greatest Number
number = [1,4,6,90,100,1,0]
reduce(lambda x,y : x if x>y else y , number) 
```

```python
#Factorial 
def fact(n) : 
    return reduce(lambda x,y : x*y, range(1,n+1))
```

```python
fact(5)
```
