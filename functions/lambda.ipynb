```python
# Lambda function : function with  lambda keyword and no def and return statement
#syntax
  # lambda argument : expression(computation)
```

```python
def square(n) : 
    return n**2
square(5)
```

```python
square_lam = lambda n : n*n
square_lam(5)
```

```python
#Even odd
is_Even = lambda num : num % 2 == 0 
is_Even(5)
```

```python
# length of string
length = lambda l : len(l)
length("Pramod")
```

```python
length(["hello" , "mohit", 1,2])
```

```python
# Normal sorting
x = ["Pramod" , "Ajay", "Rahl" , "Manikshow" , "alu"]
sorted(x)
```

```python
sorted(x,key = lambda x : len(x))
```

```python
# fibonaaci recurssive
fib = lambda  n : n if n<= 1 else fib(n-1) + fib(n-2)
[fib(i) for i in range(10)] # <-- list compreshenstion
```

```python
# Factorial
fact = lambda n : 1 if n<=1 else n*fact(n-1)
[fact(i) for i in range(10)]
```
