```python
# List Comprehesion
```

```python
prices = [10,20,30,40,50]
double_price = []
for i in prices : double_price.append(i*2)
double_price
```

```python
double_price = [price * 2 for price in prices]# this is  list comprehension : good practice : you must follow while mofiying of whole lits
double_price
```

```python
# Conditional  List comprehension
```

```python
# Conditional  List comprehension 
emails = ['ab@gmail.com' ,'qw@yahoo.com' , 'as@outlook.com' ,' mr@yahoo.com']
[email for email in emails if email.endswith('@yahoo.com')]
```

```python
#Nested list Comprehension
```

```python
pairs= []
for i in [1,2,3] : 
    for j in [4,5,6] : 
        pairs.append([i,j])
pairs
```

```python
[[i,j] for i in [1,2,3] for j in [4,5,6]]
```

```python
[i*j for i in [1,2,3] for j in [4,5,6]]
```
