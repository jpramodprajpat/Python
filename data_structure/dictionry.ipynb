```python
# Converting two list to dicitionary using : Zip
```

```python
a = ["name" , "roll" , "subjects" , "semester"]
b = ["karan" , 101 , "tally" , "1"]
```

```python
d ={}
for a,b in zip(a,b):
    d[a] = b
d
```

```python
# way -2
d1 = {}
for i in range(len(a)) :  # length must be same
    d1[a[i]] = b[i]
d1
```

```python
user_id = [1,2,3]
user_name = ["ram" , "shyam" , "rahul"]
{id : name for id , name in zip(user_id, user_name)}
```
