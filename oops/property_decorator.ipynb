```python
# Property decorator >> you are allowed to use class method as a attribute
```

```python
class Circle : 
    def __init__(self, radius): 
        self.radius = radius
    def area(self):
        radius = self.radius
        return 3.14*radius**2
```

```python
obj = Circle(5)
```

```python
obj.area()
```

```python
# Using Property decorator
class Circle : 
    def __init__(self, radius): 
        self.radius = radius
    @property
    def area(self):
        radius = self.radius
        return 3.14*radius**2
```

```python
obj = Circle(4)
```

```python
obj.area # using property decorator - no of use of parenthise ()
```

```python
class Student:
    def __init__(self, name , price):
        self.__name = name
        self.__price = 3000
    @property
    def access_price(self) :
        return self.__price
```

```python
stu = Student("MO=hot" , 2300)
```

```python
stu.access_price # so there methode is acting as argument or we also accessing the private variable :) wooow 
```

```python
# so step -1 : private variable -> accessible 
# now step -2 : try to update or modify private varible
```

```python
class Student:
    def __init__(self, name , price):
        self.__name = name
        self.__price = 3000
    @property
    def access_price(self) :
        return self.__price
    
    @access_price.setter
    def update_price(self , new_price): 
        self.__price = new_price
```

```python
stu = Student("MO=hot" , 2300)
```

```python
stu.access_price
```

```python
stu.update_price = 1122
```

```python
stu.access_price
```

```python
# step - 2 : ✅
# step-3 : delete private varible
```

```python
class Student : 
    def __init__(self, name , price): 
        self.__name = name
        self.__price = price
    
    @property
    def access_price(self): 
        return self.__price
    
    @access_price.setter
    def access_price(self , new_price):
        self.__price = new_price
    
    @access_price.deleter
    def access_price(self):
        del self.__price
```

```python
stu = Student("Pramod" , 1299)
```

```python
stu.access_price
```

```python
stu.access_price = 10000
```

```python
stu.access_price
```

```python
del stu.access_price
```

```python
stu.access_price
```
