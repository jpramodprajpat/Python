```python
# Methode Overriding >> re-writing/re-defining  methode of parent class in dervied child class
class Fruit: 
    def fruit_info(self) : print("Inisde parent class")
class Apple(Fruit) : 
    def fruit_info(self) : print("Inisde  overriding methode of parent class")
    def apple_info(self) : print("inside chiild class")
```

```python
obj = Apple()
```

```python
obj.fruit_info()
```

```python
# Methode Overriding >> child class is very powerful 
# Note : Methode Overriding alawage happens between two class
```

```python
# Multi-level Inheritance

# Parent --> child -1 ---> child-2
# child -2 will have properties of Parent and child-1
```

```python
class GrandParent : 
    def grand_info(self) : print("dada ji..")
class Parent(GrandParent) : 
    def parent_info(self) : print("papa ji...")
class Child(Parent) : 
    def child_info(self) : print("I am the King")
```

```python
ob = Child()
```

```python
ob.grand_info()
```

```python
ob.parent_info()
```

```python
ob.child_info()
```

```python
class A : 
    def method(self) : 
        print("A")
class B(A) : 
    def method(self) : 
        print("B")
class C(A) :
    def method(self) : 
        print("C")
class D(B,C) : 
    pass
```

```python
obj = D()
```

```python
obj.method() # be will get priority : becuase it came first >> Diamond Problem resolution (MRO) memeber r order
```

```python
# Abstraction : Concept hiding complexities and showing necessary

# Abstract class >> can't be Instantiated (object nahi banta abstract class ka)
# Abstract class alwage a sub class
# Abstract class : methode are just declared and not implemented
# if child class who inherited that abstaract class and not implemented all methods it will not throw an error
```

```python
import abc # collection of multiple functions in module
```

```python
class PwSkills:
    @abc.abstractmethod
    def details(self) : pass
    @abc.abstractmethod
    def subjects(self) : pass
    @abc.abstractmethod
    def marks(self) : pass
```

```python
class Student(PwSkills) : 
    def details(self) : print("this is details")
    def subjects(self) : print("this is subhecs")
```

```python
s = Student()
```

```python
s.details()
```

```python
s.subjects()
```

```python
s.marks() # will not throw an error , even it is not implemented
```
