```python
# Attribute/Properties can also be accesed like methode/functions of the class 
class Bank : 
    def __init__(self):
        self.balance = 0
    def deposite(self, amount) : 
        print(f" Money is deposited : ${amount}")
        self.balance += amount
    def withdraw(self , amount) : 
        if amount > self.balance : 
            print("Sorry you don't have sufficeint balance")
        else : 
            print(f" Withdrawed amount is : ${amount}")
            self.balance -= amount
```

```python
pramod = Bank()
pramod.balance
```

```python
pramod.deposite(1000)
```

```python

pramod.withdraw(400)
```
