```python
class Student : 
    def __init__(self,name) : # init methode >> Instance methode
        self.name = name
    
    @classmethod  # class methode >> associated to class not instance >> can be called without object creation
    def student_details(cls, name) : 
        print("hello class methode : ",name)

    def testing() : print("hello testing : " , name)
```

```python
class Scalar : 
    method = "minmax"

    @classmethod
    def set_method(cls, new_name ) : 
        cls.method = new_name  #using cls is very very important here
```

```python
class Student:
    total_stu = 0

    def __init__(self,name) : 
        self.name = name
        print("hello ", self.name)

    @classmethod
    def count(cls) :
        cls.total_stu +=1
```

```python
class Student : 
    total_stu = 0 # Class variable

    def __init__(self , name ): # Instance Methode
        self.name = name
        Student.total_stu += 1 # this is "class-varible" accessible to all the instance
    
    @classmethod  # Class Method
    def get_total_student(cls) :  return  Student.total_stu
```

```python
a = Student("a")
```

```python
Student.get_total_student()
```

```python
b = Student("b")
Student.get_total_student()
```
