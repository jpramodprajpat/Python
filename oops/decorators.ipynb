```python
# A function decorator >>in Python is a special feature that lets you "add extra behavior" to a function without changing its original code.
# Decorator >> A function that takes a another function and add something and return new funciton
```

✅ Why decorators are used?

To add common features like:

--> logging (print when a function runs)
--> timing (how much time taken)
--> authentication / permission checking
--> caching
--> validation
logging , timing, authentication , caching and validation

```python
# Function Decorator 
```

```python
# Normal function >> without decorator
def func(): 
  print("before computation")
  print(12*12)
  print("After computation")
 
```

```python
func()
```

```python
# Decorator Function -

def my_decorator(func) : # decorator function -> that takes another function as a argument  
    def wrapper() :  # wrapper - nested function >>adding functionality before and after calling the functions
        print("the line before computation")
        func()
        print("the line after computation")
    return wrapper
```

```python
@my_decorator # syntex to decorate greet() function
def greet(): 
    print("hello, good morning")
```

```python
greet()
```

```python
# Another Usecase
import time
def time_decorator(func) :
    def timer() :
        start = time.time()
        func()
        end = time.time()
        print(end-start)
    return timer
# Note : if we use return timer() it directly excecute the function even without calling        
```

```python
@time_decorator
def county() : 
    print(110000*12000)
```

```python
county()
```

```python
def ex1(func) : 
    def wrapper() : 
        print("before ")
        func()
        print("After")
    return wrapper
```

```python
@ex1
def test1() : print("this is test running ..")
```

```python
# Class Decorator
```

```python
class MyDecorator:
    def __init__(self , func):
        self.func = func
    def __call__(self):
        print("Something is happening before function.")
        self.func()
        print("Something is happening after function.")
```

```python
@MyDecorator
def greet() : 
    print("hello")
greet()
```

```python
# Example -2 > Class Decorator
```

```python
class Ex2 : 
    def __init__(self , func):
        self.func = func
    def __call__(self):
        print("Before")
        self.func()
        print("After")
```

```python
@Ex2
def test() : print("this is test running")
```

```python
test()
```

```python
# Some Built-in Decorators>>
# - @classmethod >> this methode take class itself as first argument 
```

```python
class Math : 
    @classmethod
    def add(cls , x, y): 
        return cls.__name__ , x, y # cls.__name__ : it is class name
```

```python
Math.add(3,4)
```

```python
# Class method are bound to the class ,not the instance of class
# class methode takes class as their first argument 
# Using this you can modify : class state / Behaviour
```

```python
# Static method -> another decortor     
# This is the method which can be called without creating any instance of class, and without using "self and cls"
```

```python
# Earlier
class Math: 
    def add(self,x,y):
        return x+y
```

```python
a = Math() # made an object/Instance
```

```python
a.add(2,3) # this is they way to call regular class method
```

```python
# use of static method
class Math:
    @staticmethod
    def add(x,y): # no need of cls or self
        return x+y
```

```python
Math.add(2,3) # no need of making object
```

```python
# class methode : 
 # use it when you want to change class level data

# static methode : 
  # when you want to interact with class level data
```
