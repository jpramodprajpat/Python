```python
# Polymorphism >> poly means "many" and morphism means "forms"
# Method Overriding ✅
# Method Overloading => ppython doesn't support directly>>BCZ last same named methode is considered as final methode
```

```python
class stud :
    def student(self) : 
        print("welocme guys")
    def student(self, name = "") : 
        print("welocme guys" ,name)
    def student(self , name="" , course="") : 
        print("welocme guys",name ,course)
```

```python
obj = stud()
```

```python
obj.student()
```

```python
obj.student("Pramod")
```

```python
obj.student("pramod" , "data science")
```

```python
# Methode Overloading >> student methode taking differnt forms , last methode overloads
class stud :
    def student(self , name="" , course="") : 
        print("welocme guys",name ,course)
```

```python
ob = stud()
```

```python
ob.student("Priya")
```

```python
ob.student()
```
