```python
# to make Private (variable/methode)>> use __
```

```python
class student:
    def __init__(self ,name , degree) : 
        self.name = name
        self.__degree = degree
    def show(self):
        print("student details", self.name , self.__degree)
```

```python
raju = student("raju" , "BA")
```

```python
raju.degree # this is private (can't be accessed outside class ,not even by their objects)
```

```python
raju.name
```

```python
raju.show()
```

```python
# But still you can acccess and modify it
raju._student__degree
```

```python
raju._student__degree = "phd"
```

```python
raju.show()
```

```python
# Private Method
class student:
    def __init__(self ,name , degree) : 
        self.name = name
        self.__degree = degree
    def show(self):
        print("student details", self.name , self.__degree)
    def __private_methode(self) : 
        print("this is a private methode")
    def accessing_private_methode(self) : 
        self.__private_methode()
```

```python
ob = student("bhim" , "B.com")
```

```python
ob.__private_methode()
```

```python
ob._student__private_methode()
```

```python
ob.accessing_private_methode()
```

```python
'''
In same class to access any methode/variables in any another methode : 
  self.__varName
  self.__funcName
'''
```

```python
'''
To access private Methode or variable outside through Object : 
 obj._className__privateName (weather variable or methode)
'''
```

```python
# Protected >> Not much used >> used Inherited class(  With in class and its sub class, protected memeber can access) ,(_)
```

```python
class Collage:
    def __init__(self) :
        self._collage_name = "Pavan ka collage" 
class Student(Collage) : 
    def __init__(self , name) : 
        self.name = name
        Collage.__init__(self) # <--- Accessing varible of base class >> className__init__(self)
        
    def show(self) : 
        print("name", self.name , self._collage_name)# calls directly the base class
```

```python
ob = Student("Pramod")
```

```python
ob.show()
```
