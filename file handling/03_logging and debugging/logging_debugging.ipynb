```python
# Program for separating integers and string from the list : 
import logging;
list_ele = [1, "hello" , [3,"world"] , 4, [5,"python"]];

# sol: 
logging.basicConfig(filename = 'program.log' , level = logging.DEBUG )
l_int = []
l_str = []

for i in list_ele: 
    logging.info(f"Processing each element of list : {i}");
    if type(i) == list : 
        for j in i : 
            logging.info(f"processing sublist : {j}");
            if type(j) == int : 
                l_int.append(j);
            elif type(j) == str :
                l_str.append(j);
    elif type(i) == int : 
        l_int.append(i);
    else : l_str.append(i);

logging.info(f"Result : list of integers{l_int}")
logging.info(f"Result : list of String{l_str}")
logging.shutdown()
```

```python
import logging

logging.basicConfig(
    filename="prime.log",
    level=logging.DEBUG
)

num = 12;

logging.info(f"Checking whether {num} is prime")

flag = True; # assuming number is  prime - by this it already assumes 1 and 2 as prime

for i in range(2,num) : 
    logging.debug(f"checking divisor of {i}");
    if(num % i==0):
        logging.warning(f"factor found {i}");
        flag = False;
        break;

if flag:
    logging.info(f"{num} is Prime")
    print(f"{num} is Prime")
else:
    logging.info(f"{num} is Not Prime")
    print(f"{num} is Not Prime")

logging.info("Program completed")
```
