```python
import csv
```

```python
#csv file : comma separated values file
data = [
    ["Name", "Course", "Fees"],
    ["Pramod", "Python", 1000],
    ["Suresh", "Java", 1200],
]
```

```python
with open("demo.csv" , "w") as file : 
    w = csv.writer(file);
    for i in data: 
        w.writerow(i);
```

```python
with open("demo.csv" ,"r") as file : 
    r = csv.reader(file);
    for i in r : 
        print(i);
```

```python
# Json Data 
import json
```

```python
data = {
    "Name" : "Pramod",
    "Course" : "Python",
    "Fees" : 1000
}
```

```python
with open('demo_json.json', 'w') as file:
    json.dump(data, file)
```

```python
with open('demo_json.json' , 'r') as file : 
    data = json.load(file);
    print(data);
```
