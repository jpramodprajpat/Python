```python
try :
    file = open("demo.txt" , "w");
    file.write("hello ,this is me");
except Exception as e : 
    print(e);
else :
    file.close();
finally:
    print("this is always executed")
```

```python
# Else block is executed when there is no exception in try block. It is used to close the file after writing to it.
# if try block is executed successfully then else block will be executed. If there is an exception in try block then else block will not be executed.
```
