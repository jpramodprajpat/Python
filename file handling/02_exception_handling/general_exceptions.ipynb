```python
# File not found exception
try : 
    with open("demo.txt", "r") as file : 
        data = file.read()
except FileNotFoundError : 
    print("File not found");
else : 
    print(data);
```

```python
# Permission Error
try : 
    with open("demo.txt" , "r") as file :
        data = file.read()
        print(data)
except PermissionError : 
    print("primission not granted");
```
